### Indexing in DataFrame

There are 2 types of indexing in Pandas

    1. Using numbers(record or column index) using iloc
    2. Using names(record name or column name) using loc
 
__key take-away
    1. The .loc method fetches all the required column & rows for start to end range(inclusive)
    2. If the set_index is used on int column and then .loc works on the values of this column
    3. If the set_index is used on non-int column then slicing doesn't work. It fails

### Merging, Concatenating and Appending

```python
df1= pd.DataFrame ({
    'id':[1,2,3,4,5],
    'name':['a','b','c','d','e'],
    'sub':['sub1','sub2','sub3','sub4','sub5'] 
})

df2=pd.DataFrame({
    'id':[1,2,3,4,5],
    'name':['b','c','d','e','f'],
    'sub':['sub3','sub4','sub5','sub6','sub7']
})
```

```python
print(df1)
print('\n\n')
print(df2)
```

### Concatenate two dataframes

```python
pd.concat([df1,df2],axis=0)    # concat row wise
```

```python
pd.concat([df1,df2],axis=1)    # concat column wise
        # concatenating columns side by side retains all the column names. It doesn't merge any column
```

__Merging 2 Dataframes__ This works like SQL joins - inner join and outer join

```python
pd.merge(left=df1, right=df2, on='sub',how='inner')    # common column is displayed only once
```

```python
pd.merge(left=df1,right=df2,on='sub')    # when we don't define 'how' option then by default it's a inner join
                                         # Joining 2 DataFrame using 'sub' as key
```

```python
pd.merge(left=df1,right=df2, on='sub', how='left')    # left join, nulls are represented as NaN
                                                      # left, right and outer join
```

```python
df1.merge(df2,left_on='sub', right_on='sub',how='right')
```

```python
pd.merge(left=df1, right=df2, on='sub', how='outer')      # full outer join  with nulls represented by NaN
```

```python
df3=(pd.merge(left=df1,right=df2, on='sub', how='left'))
```

### Display the data of only left table - no intersection

```python
### creating dataframe from list
import pandas as pd
Mylist1 = [(1,10),(2,20),(3,30),(4,40),(5,50)]
labels1 = ['ID','NUM']
df1_x= pd.DataFrame(Mylist1,columns=labels1)

Mylist2 = [(1,'A'),(6,'B'),(3,'C'),(8,'D'),(10,'E')]
labels2 = ['ID','ALPHA']
df2_x=pd.DataFrame(Mylist2,columns=labels2)

#print(df1_x)
#print(df2_x)

# indicator =True - will tell which row belongs to which table like right or left only through "_merge" column
df4= pd.merge(left=df1_x,right=df2_x,on='ID', how='left',indicator=True)
print('df4\n',df4)

# Query function is used tofilter all those rows satisfying the condition, provide it in quotes
# Using dropna() - this deletes all those columns that have missing values
df3_left= pd.merge(left=df1_x,right=df2_x,on='ID', how='left',indicator=True).query('_merge=="left_only"').dropna(axis=1)

# using .query and drop(column,axis)
df4_left= pd.merge(left=df1_x,right=df2_x,on='ID', how='left',indicator=True).query('_merge=="left_only"').drop('_merge',axis=1)

# dataframe.<column-name>.notnull() function- allows to return all the rows with non null values of that column
#df3 = df3_left[~df3_left.ALPHA.notnull()]        ## fetch all those rows where there is no
           
print("df3_left \n",df3_left)
print("df4_left \n",df4_left)
```
